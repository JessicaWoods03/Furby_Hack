```python
import spacy

# Load spaCy's English model
nlp = spacy.load("en_core_web_sm")

# Define a function to classify the relationship
def classify_relationship(sentence):
    # Basic rules to classify the relationship type based on sentence keywords
    if any(keyword in sentence.lower() for keyword in ['on', 'at', 'near', 'by', 'located', 'south', 'east', 'west', 'north', 'northwest', 'northeast', 'southwest', 'southeast']):
        return "Geographic"
    elif any(keyword in sentence.lower() for keyword in ['larger','largest', 'smallest', 'smaller','similar', 'more', 'less', 'better', 'compared']):
        return "Comparison"
    elif any(keyword in sentence.lower() for keyword in ['is', 'has','was', 'includes', 'offers', 'provides']):
        return "Functional"
    
    return "Unknown"

# Define the function to extract parent and children using spaCy
def extract_parent_child(sentence):
    # Process the sentence with spaCy
    doc = nlp(sentence)

    parent = None
    children = []

    # Identify the parent (subject of the sentence)
    for token in doc:
        if "subj" in token.dep_:  # Subject of the sentence
            parent = token.text

        # Identify children (objects, prepositional objects, or attributes)
        if token.dep_ in ("dobj", "pobj", "attr"):  # Direct Object, Prepositional Object
            children.append(token.text)
    for ent in doc.ents:
        if ent.label_ == "DATE":  # Look cases like 17th Century
            children.append(ent.text)
    # the testing worked...children are a bit excessive in some cases 
    
    return parent, children
df['parent_node'] = None
df['child_node'] = None
# testing 
# sentence = "Nørresundby is connected to Aalborg by a road bridge."

# now for the whole dataframe to go threw
for index, row in df.iterrows():
    # Extract parent and children using the previously defined function
    parent, children = extract_parent_child(row['sentence_restructured'])
    
    # Assign the parent and child nodes to the respective columns in the current row
    df.at[index, 'parent_node'] = parent
    df.at[index, 'child_node'] = ', '.join(children)  # Join the children list if there are multiple
    
    # Classify relationship (Make sure classify_relationship function exists)
    relationship = classify_relationship(row['sentence_restructured'])
    df.at[index, 'relationship'] = relationship

# Output the results
display(df[['sentence_restructured','parent_node', 'child_node', 'relationship']])
```

```python
# Function to build a graph from all sentences in a paragraph
def build_parent_child_structure(df):
    # Initialize an empty dictionary to hold the parent-child structure
    parent_child_dict = {}

    # Group by 'content' (paragraph)
    for content, group in df.groupby('content'):
        parent_child_structure = {}

        # Iterate over each row within the paragraph group
        for _, row in group.iterrows():
            parent = row['parent_node']
            children = row['child_node'].split(', ')  # Split children if there are multiple
            relationship = row['relationship']

            # Initialize the parent node if not already in the dictionary
            if parent not in parent_child_structure:
                parent_child_structure[parent] = {}

            # Initialize the relationship type if not already in the dictionary
            if relationship not in parent_child_structure[parent]:
                parent_child_structure[parent][relationship] = []

            # Append children to the corresponding relationship type
            parent_child_structure[parent][relationship].extend(children)

        # Store the parent-child structure for this paragraph
        parent_child_dict[content] = parent_child_structure

    return parent_child_dict

# Get the structured metadata for each paragraph
parent_child_structures = build_parent_child_structure(df)

# Print the structured data (this would be inserted into ChromaDB)
for content, parent_child_structure in parent_child_structures.items():
    
    print(f"Parent-Child Structure for this paragraph:")
    print({
        "content": content,
        "parent_child_structure": parent_child_structure
    })
    print("\n") 
```

```python
# Node Graph visualizations

import networkx as nx
import matplotlib.pyplot as plt

# Initialize a directed graph
G = nx.DiGraph()
df_grouped_nodes['ontology'] = None
# Define the DataFrame (df_grouped_nodes) with your nodes and relationships
def classify_relationship_paragraphs(sentence):
    # Basic rules to classify the relationship type based on sentence keywords
    if any(keyword in sentence.lower() for keyword in ['on', 'at', 'near', 'by', 'located', 'south', 'east', 'west', 'north', 'northwest', 'northeast', 'southwest', 'southeast']):
        return "Geographic"
    elif any(keyword in sentence.lower() for keyword in ['larger','largest', 'smallest', 'smaller','similar', 'more', 'less', 'better', 'compared']):
        return "Comparison"
    elif any(keyword in sentence.lower() for keyword in ['is', 'has','was', 'includes', 'offers', 'provides']):
        return "Functional"
    
    return "Unknown"
ontology = {}
# Iterate over the rows of df_grouped_nodes (assuming each row is a paragraph)
for index, row in df_grouped_nodes.iterrows():
    # Initialize a directed graph for this paragraph
    G = nx.DiGraph()

    # Split the paragraph into sentences
    sentences = row['sentence_restructured'].split('.')
    
    # Process each sentence individually
    for sentence in sentences:
        if sentence.strip():  # Skip empty sentences
            # Extract parent and child nodes for the current sentence
            parent, children = extract_parent_child(sentence)
            
            if parent:  # Ensure there's a parent
                for child in children:
                    # Classify the relationship for this individual sentence
                    relationship = classify_relationship(sentence)
                    
                    G.add_edge(parent, child, relationship=relationship)
                    if parent not in ontology:
                        ontology[parent] = {}
                    ontology[parent][child] = relationship
                    
    df_grouped_nodes.at[index, 'ontology'] = [ontology]               
                    
                    
    
    # Visualize the graph for this paragraph
    plt.figure(figsize=(10, 10))
    pos = nx.spring_layout(G)  # Layout for visualization

    # Draw the graph with labels
    nx.draw_networkx(G, pos, with_labels=True, node_size=3000, node_color='skyblue', font_size=10, font_weight='bold')

    # Draw the edge labels (relationship types)
    edge_labels = nx.get_edge_attributes(G, 'relationship')
    nx.draw_networkx_edge_labels(G, pos, edge_labels=edge_labels)

    # Add a title with the paragraph title (e.g., Aalborg)
    plt.title(f"Node Graph for Paragraph: {row['title']}")
    plt.show()
    
```
